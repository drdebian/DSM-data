# PV: PVGIS

## Source

- Link: [Photovoltaic Geographical Information System (PVGIS)](https://re.jrc.ec.europa.eu/pvg_tools/en/tools.html): [users manual](https://ec.europa.eu/jrc/en/PVGIS/docs/usermanual), [hourly-radiation](https://ec.europa.eu/jrc/en/PVGIS/tools/hourly-radiation)
- Location and settings: ![](../../../data/PVGIS/DSMDUFlex/Screenshot_20201030_103420.png)
- See the legend at the end of the `*_orig.csv` file:

```
P: PV system power (W)
G(i): Global irradiance on the inclined plane (plane of the array) (W/m2)
H_sun: Sun height (degree)
T2m: 2-m air temperature (degree Celsius)
WS10m: 10-m total wind speed (m/s)
Int: 1 means solar radiation values are reconstructed
PVGIS (c) European Communities, 2001-2020
```

- available time frame: up to and including 2016
- sampling interval: 1 hour
- units: W
- location: Konstanz
- [legal notice](https://ec.europa.eu/info/legal-notice_en): "... reuse is allowed, provided appropriate credit is given and changes are indicated."

## Data

[Python strftime() and strptime() Format Codes](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes)

````python
my_path = "/home/kr/forschung/data/PVGIS/DSMDUFlex/"
my_file = "Timeseries_47.666_9.179_SA_1kWp_crystSi_14_36deg_1deg_2015_2016.csv"
````

````python
my_parser = lambda date: pd.to_datetime(date, format='%Y%m%d:%H%M')
df = pd.read_csv(my_path + my_file, 
                 index_col=0, skiprows=10, 
                 parse_dates=True, date_parser=my_parser)
df.head(2)
````

````python
df.tail(2)
````

````python
# resample to dt = 0.25 h using linear interpolation:

dt = 0.25  # sampling interval in hours
df = df.resample(rule='15Min').mean().interpolate(method='linear')
if 0: # check: 
    df.loc['2016-07-01 12:00':'2016-07-01 13:00','P'].plot(marker='o')
    plt.grid(True)
    df.loc['2016-07-01 12:00':'2016-07-01 13:00',]
````

````python
# exptrapolate December 31 23:00:

add_df = pd.DataFrame(0.0, columns=['P'], 
                      index = pd.date_range(
                      start='2016-12-31 23:15:00', 
                      end  ='2016-12-31 23:45:00', freq='15Min') 
                     )
df = df.append(add_df)
````

````python
df.tail(4)
````

````python
# rescale to kW:

df['P'] = df['P']/1000
````

````python
# delete unneeded columns:

df.drop(['G(i)', 'H_sun', 'T2m', 'WS10m', 'Int'], axis=1, inplace=True)
````

````python
# annual energy generations:

print(f"energy generation in 2015: {df.loc['2015','P'].sum()*dt:.2f} kWh")
print(f"energy generation in 2016: {df.loc['2016','P'].sum()*dt:.2f} kWh")
````

````python
# qgrid.show_grid(df.loc['2016'])
````

````python
# predictions: function and example of 1 day prediction

my_day_str = "2016-03-01"

my_format ="%Y-%m-%d"
my_day_ts = pd.to_datetime(my_day_str, format=my_format)

def predict_pv(df, day_ts, p=0.8):
    """
    p ... weight for day
    """
    my_one_day_before_ts = day_ts - pd.Timedelta(1, unit='days')
    my_prediction = df.loc[my_one_day_before_ts.strftime(my_format),'P'].values
    my_prediction = (1 - p)*my_prediction + p*df.loc[my_day_ts.strftime(my_format),'P'].values
    return my_prediction

my_prediction = predict_pv(df, my_day_ts)
````

````python
plt.figure(figsize=(10,6))
my_times = df.loc[my_day_str,'P'].index
my_true_values = df.loc[my_day_str,'P'].values
plt.plot(my_times, my_true_values, '.-', label='true')
plt.plot(my_times, my_prediction, '.-', label='prediction')
plt.legend()
plt.grid()
````

````python
# add column 'P_pred' with 2016 daily predictions:

df['P_pred'] = np.NaN*df['P']
for my_day_ts in pd.date_range(start='2016-01-01', end='2016-12-31'):
    my_prediction = predict_pv(df, my_day_ts)
    df.loc[my_day_ts.strftime(my_format),'P_pred'] = my_prediction
````

````python
df.loc['2016-4-02'].plot(figsize=(10,6), grid=True, marker='.');
````

````python
# export to pickle:

df.to_pickle('../data/pv_df.pickle')  # maybe set protocol=4 later
````

# Household: OPSD

[data.open-power-system-data.org](https://data.open-power-system-data.org/household_data/): This data package contains measured time series data for several small businesses and residential households relevant for household- or low-voltage-level power system modeling.
All measurement devices provided **cumulative energy** consumption/generation over time. 
Hence overall energy consumption/generation is retained, in case of data gaps due to communication problems.
All data gaps are either interpolated linearly, or filled with data of prior days.

- DE_KN_residential1_grid_import: number (float), Energy imported from the public grid in a residential building, located in the suburban area in kWh
- DE_KN_residential1_pv: number (float), Total Photovoltaic energy generation in a residential building, located in the suburban area in **kWh** 

## Data

````python
my_path = "/home/kr/forschung/data/OPSD/Household_Data/"
my_file = "household_data_15min_singleindex_filtered.csv"
````

````python
if 0:   # read raw data from csv
    # 2014-12-11T17:45:00Z,2014-12-11T18:45:00+0100
    # https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes
    my_parser = lambda date: pd.to_datetime(date, format='%Y-%m-%dT%H:%M:%S%z').tz_localize(None)
    
    df = pd.read_csv(my_path + my_file,
                     # nrows=10000,         # later None or commented
                     usecols=['cet_cest_timestamp',
                              'DE_KN_residential1_grid_import', 'DE_KN_residential1_pv',
                              'DE_KN_residential2_grid_import', 
                              'DE_KN_residential3_grid_import', 'DE_KN_residential3_pv',
                              'DE_KN_residential4_grid_import', 'DE_KN_residential4_pv',
                              'DE_KN_residential5_grid_import', 
                              'DE_KN_residential6_grid_import', 'DE_KN_residential6_pv'],
                     parse_dates=['cet_cest_timestamp'], # makes these columns to type Timestamp
                     date_parser=my_parser,
                     index_col='cet_cest_timestamp'
                    )
    # for i, item in enumerate(df.columns):
    #     print(f"{i}: {item}")
    df.index.name = 'time'
    # export raw data to pickle:
    df.to_pickle('../data/hh_raw_df.pickle')  # maybe set protocol=4 later
else:    # import pickle:
    df = pd.read_pickle('../data/hh_raw_df.pickle')
````
